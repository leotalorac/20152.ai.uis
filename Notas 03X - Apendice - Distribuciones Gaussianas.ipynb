```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Understand multiplication and addition of Gaussians

```python
def multiply_gaussians(mu1, std1, mu2, std2):
    std1 = std1 if std1!=0 else 1.e-80
    std2 = std2 if std2!=0 else 1.e-80
    mean = (mu2*std1**2 + mu1*std2**2) / (std1**2+std2**2)
    std  = np.sqrt(1. / (1./(std1**2) + 1./(std2**2)))
    return (mean, std)

def add_gaussians(mu1, std1, mu2, std2):
    mean = mu1 + mu2
    std  = np.sqrt(std1**2 + std2**2)
    return mean, std
```

```python
from scipy.stats import norm
mu1, std1 = 5., np.sqrt(1)
mu2, std2 = 2., np.sqrt(2)
mmu, mstd = multiply_gaussians(mu1,std1, mu2,std2)
amu, astd = add_gaussians(mu1,std1,mu2,std2)
print mmu, mstd
print amu, astd
x = np.linspace(0,11,100)
fig = plt.figure(figsize=(10,3))
fig.add_subplot(121)
plt.plot(x, norm(mu1, std1).pdf(x), label="p1")
plt.plot(x, norm(mu2, std2).pdf(x), label="p2")
plt.plot(x, norm(mmu, mstd).pdf(x), label="p1*p2", lw=6, alpha=0.5)
plt.ylim(0,0.6)
plt.legend()
fig.add_subplot(122)
plt.plot(x, norm(mu1, std1).pdf(x), label="p1")
plt.plot(x, norm(mu2, std2).pdf(x), label="p2")
plt.plot(x, norm(amu, astd).pdf(x), label="p1+p2", lw=6, alpha=0.5)
plt.ylim(0,0.6)
plt.legend()
```
